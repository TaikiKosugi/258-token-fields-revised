```python
!pip install selenium
!pip install pandas

# ChromeDriverのバージョンを取得
!apt-get update
!apt install -y wget unzip
!wget https://dl.google.com/linux/direct/google-chrome-stable_current_amd64.deb
!dpkg -i google-chrome-stable_current_amd64.deb
!apt --fix-broken install -y
!apt install -y chromium-chromedriver

# ChromeDriverのパスをColab環境に設定
import os
os.environ["PATH"] += ":/usr/lib/chromium-browser/"

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.action_chains import ActionChains # Import ActionChains
from selenium.webdriver.support.ui import Select # Import the Select class
import pandas as pd


# Chromeオプションを設定
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# ブラウザを起動
driver = webdriver.Chrome(options=options) # 'chromedriver'を削除

# ログイン情報を設定
login_url = "https://admin.japanstreet.jetro.go.jp/"
username = "taiki_kosugi"  # 実際のユーザー名に置き換えてください
password = "1410hand1410"  # 実際のパスワードに置き換えてください

try:
    # ログインページにアクセス
    driver.get(login_url)

    # IDフィールドを見つけて入力
    id_field = driver.find_element(By.NAME, "login[login]")
    id_field.send_keys(username)

    # パスワードフィールドを見つけて入力
    password_field = driver.find_element(By.NAME, "login[password]")
    password_field.send_keys(password)

    # ログインボタンをクリック
    login_button = driver.find_element(By.NAME, "commit")
    login_button.click()

    # ログイン後のページが読み込まれるまで待機
    time.sleep(5)

    # <li class="top user">要素を見つけてホバー
    top_user_element = driver.find_element(By.CSS_SELECTOR, "li.top.user")
    ActionChains(driver).move_to_element(top_user_element).perform()

    # サブメニューが表示されるまで待機
    time.sleep(1)

    # 「フロントユーザ検索」リンクを見つけてクリック
    front_user_search_link = driver.find_element(By.LINK_TEXT, "フロントユーザ検索")
    front_user_search_link.click()

    # リンクをクリック後のページを表示
    time.sleep(5)

    # ドロップダウンを見つけてオプションを選択
    select_element = Select(driver.find_element(By.ID, "condition_disp_count"))
    select_element.select_by_value("20")

    # 「検索」ボタンをクリック
    search_button = driver.find_element(By.NAME, "commit")
    search_button.click()

    time.sleep(5)

    print(driver.page_source)


    # driver.page_sourceから<tr>の数をカウント
    import re

    html_source = driver.page_source
    tr_count = len(re.findall(r"<tr>", html_source))

    print("<tr> count:", tr_count)

    # table要素を取得
    tables = driver.find_elements(By.TAG_NAME, 'table')
    table = tables[1]

    # tableの行を取得
    rows = table.find_elements(By.TAG_NAME, 'tr')
    print("number of rows: " + str(len(rows)))

    # データを格納するリスト
    data = []

    # 各行からデータを取得
    for row in rows[1:]:  # ヘッダー行をスキップ
        cells = row.find_elements(By.TAG_NAME, 'td')
        print("number of cells: " + str(len(cells)))

        row_data = [cell.text for cell in cells]

        # 必要なデータのみ抽出
        extracted_data = {
            "No.": row_data[0],
            "サイト": row_data[0],
            "ユーザ名": row_data[2],
            "リンク": cells[2].find_element(By.TAG_NAME, 'a').get_attribute('href') if cells[2].find_elements(By.TAG_NAME, 'a') else None,  # リンクURLを取得
            "バイヤー名": row_data[3],
            "住所": row_data[4],
            "電話番号": row_data[5],
            "顧客区分": row_data[6],
            "登録日": row_data[7],
            "履歴": row_data[8],
            "操作": row_data[9]
        }
        data.append(extracted_data)

        print(extracted_data)
        # print(row_data)
        # data.append(row_data)

    # pandas DataFrameを作成
    df = pd.DataFrame(data)

    # JSONファイルとして保存
    df.to_json("table_data.json", force_ascii=False, orient='records')

    # print(driver.page_source)

    # import json

    # with open('table_data.json', 'r') as f:
    #     data = json.load(f)

    # print(data)

finally:
    # ブラウザを閉じる
    driver.quit()
```
